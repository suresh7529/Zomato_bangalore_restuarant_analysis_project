## IMPORT LIBRARIES

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
plt.style.use('dark_background')
```

```python
dat = pd.read_csv('zomato.csv')
```

```python
dat.columns
```

## Removing unwanted columns from the dataset

```python
dat = dat.drop(['url','address','phone','menu_item','dish_liked','reviews_list'], axis = 1)
```

```python
dat
```

```python
dat.head()
```

```python
dat.info()
```

## Dropping Duplicates values

```python
dat.drop_duplicates(inplace = True)
```

```python
dat.shape
```

## Cleaning the rate column

```python
dat.head()
```

```python
dat['rate'].unique()
```

## Removing 'NEW','-',and '/5' from rate column

```python
def handlerate(value):
    if(value=='NEW' or value=='-'):
        return np.nan
    else:
        value = str(value).split('/')
        value = value[0]
        return float(value)

dat['rate'] = dat['rate'].apply(handlerate)
dat['rate'].head()
```

```python
dat.rate.isnull().sum()
```

## Filling Null Values in Rate Column With Mean 

```python
#data['rate'].fillna(data['rate'].mean(), inplace = True)
dat.fillna({'rate': dat['rate'].mean()}, inplace=True)
dat.rate.isnull().sum()
```

```python
dat.info()
```

## Dropping Null Vlaues

```python
dat.dropna(inplace = True)
dat.head()
```

```python
dat.rename(columns = {'approx_cost(for two people)':'Cost2People','listed_in(type)':'Type'}, inplace = True)
```

```python
dat.head()
```

```python
dat['location'].unique()
```

```python
dat['listed_in(city)'].unique()
```

## listed_in(city) and location both are there,lets keep only one

```python
dat = dat.drop(['listed_in(city)'], axis = 1)
dat.head()
```

```python
dat['Cost2People'].unique()
```

## Removing ',' from the Cost2People column 

```python
def handlecomma(value):
    value = str(value)
    if ',' in value:
        value = value.replace(',', '')
        return float(value)
    else:
        return float(value)

dat['Cost2People'] = dat['Cost2People'].apply(handlecomma)
dat['Cost2People'].unique()
```

```python
dat.head()
```

## Cleaning Res_Type column

```python
dat['rest_type'].unique()
```

```python
dat['rest_type'].value_counts()
```

```python
rest_types = dat['rest_type'].value_counts(ascending = False)
rest_types
```

```python
rest_types_lessthan1000 = rest_types[rest_types<1000]
rest_types_lessthan1000 
```

## Makeing Rest Type less than 1000 in frequency as others

```python
def handle_rest_type(value):
    if(value in rest_types_lessthan1000):
        return 'others'
    else:
        return value

dat['rest_type'] = dat['rest_type'].apply(handle_rest_type)
dat['rest_type'].value_counts()
```

```python
dat.head()
```

## Cleaning location column

```python
dat['location'].value_counts()
```

```python
location = dat['location'].value_counts(ascending = False)
location_lessthan300 = location[location<300]
```

```python
def handle_location(value):
    if(value in location_lessthan300):
        return 'orthers'
    else:
        return value

dat['location'] = dat['location'].apply(handle_location)
dat['location'].value_counts()
```

## Cleaning cuosines column

```python
dat.head()
```

```python
dat['cuisines'].value_counts()
```

```python
cuisines = dat['cuisines'].value_counts(ascending = False)
cuisines_lessthan100 = cuisines[cuisines<100]
```

```python
def handle_cuisines(value):
    if(value in cuisines_lessthan100):
        return 'orthers'
    else:
        return value

dat['cuisines'] = dat['cuisines'].apply(handle_cuisines)
dat['cuisines'].value_counts()
```

```python
dat['Type'].value_counts()
```

## Now the data is cleaned ,let's jump to visualization

## 1. Count plot of various location

## Summary:


The countplot shows the number of restaurants available in each location of Bangalore.
From the visualization, we observe that certain areas — such as BTM Layout, Koramangala 5th Block, and Indiranagar — have the highest concentration of restaurants.
These locations are well-known for their vibrant food culture, dense population, and active nightlife, attracting both dine-in and online customers.
In contrast, suburban or residential areas show a smaller number of restaurants, indicating lower commercial activity or limited customer base.
